```python
import cv2
import matplotlib.pyplot as plt
from glob import glob
import numpy as np
from math import ceil
```

```python
def plot_images(images: list, positions: np.array = None):
    n_images = len(images)
    n_columns = 7
    n_rows = ceil(n_images/n_columns)

    fig, axes = plt.subplots(n_rows, n_columns, figsize=(10, 9), sharex=True, sharey=True)

    plt.subplots_adjust(left=0.0,
                        bottom=0.0,
                        right=1,
                        top=1,
                        wspace=0.05,
                        hspace=0)
    
    ax = axes.flatten()

    for axs in ax[n_images:]:
        axs.remove()

    L = 400
    for i, img in enumerate(images):
        ax[i].set_axis_off()
        ax[i].imshow(img, 'gray')

        if positions is not None:
            x = positions[i][0]
            y = positions[i][1]

            rect = plt.Rectangle(positions[i]-L//2, L, L,ec="red", fc='none')
            ax[i].add_patch(rect)
            ax[i].plot(positions[:i+1, 0], positions[:i+1, 1],'-o', alpha=0.5, ms=1, c='r')


    
    plt.show()
```

```python
imgs_dir = '../cogum_images/*'
imgs_paths = glob(imgs_dir)
imgs_paths = sorted(imgs_paths)

# Create list with all images in gray scale
images_gray = [ cv2.imread(image_path, cv2.IMREAD_GRAYSCALE) for image_path in imgs_paths]
images_gray = images_gray[:42]
# Visualizando imagens
plot_images(images_gray)
```

```python
from skimage import feature
images_canny = [ 
             feature.canny(img, sigma=3)*255.0
             for img in images_gray]

plot_images(images_canny)
```
